```python
import gensim
import jieba
import glob 
import os
import nltk
import logging
from gensim.models import Phrases
from gensim.corpora import Dictionary

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
corpus = []
for p in glob.glob('/Users/davidzhang/Downloads/data/*.txt'):
    name = os.path.basename(p)
    idx = name.split('-')[0]
    author = name.split('-')[1]
    title = name.split('-')[-1]
    title = list(jieba.cut(title))
    content = open(p).read()
    content = list(jieba.cut(content))
    corpus.append({'id':idx, 'author':author, 'title':title, 'content':content})
```

```python
bigrams = Phrases([x['content'] for x in corpus], min_count=20)
```
